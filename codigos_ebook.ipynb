M√©dia (mean):

---



```python
import numpy as np

data = np.array([10, 20, 30, 40, 50])
mean_value = np.mean(data)
print(f"M√©dia: {mean_value:.2f}")
```

Desvio padr√£o (standard deviation):

---



```python
# Calculando o desvio padr√£o
std_value = np.std(data)
print(f"Desvio Padr√£o: {std_value:.2f}")
```

Valor m√≠nimo (minimum) e valor m√°ximo (maximum):


---



```python
# Encontrando o valor m√≠nimo e m√°ximo
min_value = np.amin(data)
max_value = np.amax(data)
print(f"Valor M√≠nimo: {min_value}, Valor M√°ximo: {max_value}")
```

Mediana (median):

---



```python
# Criando um array com valores √≠mpares
odd_data = np.array([1, 3 ,5 ,7 ,9])
# Calculando a mediana
median_odd = np.median(odd_data)
print(f"Mediana (√≠mpar): {median_odd}")

# Criando um array com valores pares
even_data = np.array([2,4 ,6 ,8 ,10])
# Calculando a mediana
median_even = np.median(even_data)
print(f"Mediana (par): {median_even}")
```

**Cap√≠tulo 7 - SciPy**

---



 Resolvendo uma equa√ß√£o diferencial ordin√°ria (ODE) com
scipy.integrate.odeint, utilizando NumPy Matplotlib e
plotando o resultado:


---



```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Definindo a fun√ß√£o que representa a ODE
def model(y, t):
    k = 0.3
    dydt = -k * y
    return dydt

# Condi√ß√£o inicial
y0 = 5

# Tempo
t = np.linspace(0, 20)

# Resolvendo a ODE
y = odeint(model, y0, t)

# Plotando o resultado
plt.plot(t, y)
plt.xlabel('Tempo')
plt.ylabel('y(t)')
plt.show()
```

Encontrando o m√≠nimo de uma fun√ß√£o com scipy.optimize.minimize e imprimindo o resultado:

---


```python
from scipy.optimize import minimize

# Definindo a fun√ß√£o
def func(x):
    return x**2 + x + 2

# Valor inicial
x0 = 0

# Encontrando o m√≠nimo
res = minimize(func, x0)

# Imprimindo o resultado
print("O m√≠nimo da fun√ß√£o ocorre em x =", res.x)
```

Realizando uma Transformada R√°pida de Fourier (FFT) com scipy.fft.fft e

---

plotando o espectro de frequ√™ncia:

```python
from scipy.fft import fft
import matplotlib.pyplot as plt

# Sinal
x = np.array([0, 1, 2, 3, 4, 5, 6, 7])

# FFT
X = fft(x)

# Plotando o espectro de frequ√™ncia
plt.stem(np.abs(X))
plt.xlabel('Frequ√™ncia')
plt.ylabel('Magnitude')
plt.show()
```
